```python
mean_squaer_feet = GS["SQUARE FEET"].mean()
mean_squaer_feet
```

```python
large_stores = GS[GS["SQUARE FEET"] >= 13000]
large_stores.head()
```

```python
# GS["WARD"]
```

```python
large_WARD_stores = GS[GS["WARD"] >= 25]
large_WARD_stores_size = large_WARD_stores["SQUARE FEET"].mean()
small_WARD_stores = GS[GS["WARD"] < 25]
small_WARD_stores_size = small_WARD_stores["SQUARE FEET"].mean()

print("Average store size of large WARD stores:", large_WARD_stores_size)
print("Average store size of small WARD stores:", small_WARD_stores_size)
```

```python
proportion_in_large_WARD_stores = large_WARD_stores.size/GS.size
proportion_in_small_WARD_stores = small_WARD_stores.size/GS.size

print("Proportion in large WARD stores:", proportion_in_large_WARD_stores)
print("Proportion in small WARD stores:", proportion_in_small_WARD_stores)
```

```python
even_WARD_stores = GS[GS["WARD"]%2 == 0]
num_even_WARD_stores = even_WARD_stores.size
num_even_WARD_stores
```
