```python
import pandas as pd
from urllib.request import urlopen 
import re
import numpy as np
%matplotlib inline
```

```python
les_miserables_url = 'http://www.gutenberg.org/files/135/135-0.txt'
les_miserables_text = urlopen(les_miserables_url).read().decode()
```

## By VOLUME

```python
les_miserables_text = re.sub('\\s+', ' ', les_miserables_text)
# volumes = les_miserables_text.split('VOLUME ')[7:]
# volumes = les_miserables_text.split('VOLUME I.--')
volumes = re.split(r"VOLUME .+?--", les_miserables_text)

# re.split(r"[,;].+?--", "a,7--bc;0--def")
```

```python
# volumes[0:5]
volumes[0]
```

```python
# Table().with_column('Text', chapters)
df = pd.DataFrame(volumes, columns = ['volume'])
df
```

## By CHAPTER

```python
les_miserables_text_by_chapter = re.sub('\\s+', ' ', les_miserables_text)
chapters = re.split(r"CHAPTER .+?--", les_miserables_text)[365:]

# les_miserables_text = re.sub('\\s+', ' ', les_miserables_text)
# volumes = les_miserables_text.split('VOLUME ')[7:]
# volumes = les_miserables_text.split('VOLUME I.--')
# volumes = re.split(r"VOLUME .+?--", les_miserables_text)

# re.split(r"[,;].+?--", "a,7--bc;0--def")
```

```python
chapters[0]
```

```python
chapters_df = pd.DataFrame(chapters, columns = ['chapter'])
chapters_df
```

```python
np.char.count(chapters, 'Bishop')
```

```python
# Major Characters
# Jean Valjean, Javert, Fantine, Cosette, Marius Pontmercy, Éponine, Grantaire
# Monsieur Thénardier and Madame Thénardier, Enjolras, Gavroche, Bishop Myriel
```

```python
Major_Characters_chapters_df = pd.DataFrame({
    'Bishop': np.char.count(chapters, 'Bishop'),
    'Valjean': np.char.count(chapters, 'Valjean'),
    'Javert': np.char.count(chapters, 'Javert'),
    'Fantine': np.char.count(chapters, 'Fantine'),
    'Cosette': np.char.count(chapters, 'Cosette'),
    'Marius': np.char.count(chapters, 'Marius'),
    'Éponine': np.char.count(chapters, 'Éponine'),
    'Grantaire': np.char.count(chapters, 'Grantaire'),
    'Thénardier': np.char.count(chapters, 'Thénardier'),
    'Enjolras': np.char.count(chapters, 'Enjolras'),
    'Gavroche': np.char.count(chapters, 'Gavroche')
})
```

```python
Major_Characters_chapters_df
```

```python
Major_Characters_chapters_df.cumsum().plot(kind='line')
```

```python
chapter_length_chapters_df = pd.DataFrame({
    'Chapter Length': [len(c) for c in chapters],
    'Number of Periods': np.char.count(chapters, '.')
}).plot(kind='scatter', x = 'Number of Periods', y = 'Chapter Length')
```
